# Solution

(1.1) 

Let g(z) be  $g(z)= \dfrac{1}{1+exp(-z)}$

Then we can also rewrite as $g(z) =(1+exp(-z))^{-1}$
Let $g(z) = [h(z)]^{-1}$
Then $\dfrac{dh(z)}{dz} = -exp^{-z}$ ; $\dfrac{dg(z)}{dh} = \dfrac{dg(z)}{dh} = -1[h(z)]^{-2}$  

$\dfrac{dg(z)}{dz} =\dfrac{dg(z)}{dh(z)}.\dfrac{dh(z)}{dz}$

=$-exp^{-z} . -1[h(z)]^{-2}$

=$\dfrac{exp^{-z}}{[h(z)]^{-2}}$

Therefore : 
$\dfrac{dg(z)}{dz}=\dfrac{exp^{-z}}{[1+exp(-z)]^{-2}}$

Which can be further rewritten as ⁉

$\dfrac{dg(z)}{dz} = g(z) (1-g(z)$

\end{align}
